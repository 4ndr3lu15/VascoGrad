```python
from vascograd import nn
import vascograd.untitled22 as interface
```

```python
import math
import random
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# make up a dataset

from sklearn.datasets import make_moons, make_blobs
X, y = make_moons(n_samples=100, noise=0.1)

y = y*2 - 1 # make y be -1 or 1
# visualize in 2D
plt.figure(figsize=(5,5))
plt.scatter(X[:,0], X[:,1], c=y, s=20, cmap='jet')
```

```python
model = nn.MLP(2, [5, 5, 1])
```

```python
def lf(y, y_hat):
    return sum([(yout - ygt)**2 for ygt, yout in zip(y, y_hat)])
```

```python
data = interface.Data(X, y)
learner = interface.Learner(model)
evaluator = interface.Evaluator(lf)
trainer = interface.Trainer(data, learner, evaluator)
```

```python
trainer.run(1000)
```
